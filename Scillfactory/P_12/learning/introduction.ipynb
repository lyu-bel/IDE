```python
import pandas as pd

melb_df_origin = pd.read_csv('data/melb_data_fe.csv')
melb_df = melb_df_origin.copy()
melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst=True)
quarter = melb_df['Date'].dt.quarter # преообразуем дату в номер квартала 

print(quarter.value_counts().iloc[1])  # сколько было продано объектов во втором по популярности продаж квартале
melb_df.columns
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car'] 
max_unique_count = 150 
for col in melb_df.columns: 
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude: 
        melb_df[col] = melb_df[col].astype('category')
#melb_df.info()
melb_df.sort_values(by='Price').head(10)
melb_df.sort_values(by='Date', ascending=False).head(10)
melb_df.sort_values(by=['Price', 'Distance']).loc[::10, ['Distance', 'Price']]
mask1 = melb_df['AreaRatio'] < -0.8
mask2 = melb_df['Type'] == 'townhouse'
mask3 = melb_df['SellerG'] == 'McGrath'
melb_df[mask1 & mask2 & mask3].sort_values(
    by=['Date', 'AreaRatio'],
    ascending=[True, False],
    ignore_index=True
).loc[:, ['Date', 'AreaRatio']]
area = melb_df.copy()
area.sort_values(by='AreaRatio', ascending=False, ignore_index=True).loc[1558, 'BuildingArea']
#melb_df.sort_values(by='AreaRatio', ignore_index=True, ascending=False).loc[1558, 'BuildingArea']
area1 = melb_df.copy()
mask1 = melb_df['Rooms'] > 2
mask2 = melb_df['Type'] == 'townhouse'
area1[mask1 & mask2].sort_values(by=['Rooms', 'MeanRoomsSquare'], ascending=[True, False], ignore_index=True).loc[18, 'Price']
melb_df.groupby(by='Type').mean()
melb_df.groupby('Type')['Price'].mean()
melb_df.groupby('Regionname')['Distance'].min().sort_values(ascending=False)
melb_df.groupby('Rooms')['Price'].agg(['mean']).sort_values(by='mean', ascending=False)
melb_df.groupby('Regionname')['Lattitude'].agg(['std']).sort_values(by='std', ascending=True)
#melb_df.columns
melb_df['Date'] = pd.to_datetime(melb_df['Date'], dayfirst=True)
melb_df.groupby('SellerG')['Price'].agg(['sum']).sort_values(by='sum', ascending=True)
date1 = pd.to_datetime('2017-05-01')
date2 = pd.to_datetime('2017-09-01')
mask = (date1 <= melb_df['Date']) & (melb_df['Date']<= date2)
melb_df[mask].groupby('SellerG')['Price'].sum().sort_values(ascending=True)
melb_df.groupby('MonthSale')['Price'].agg('describe')
m = melb_df.groupby('Regionname')['SellerG'].agg(
    		['nunique', set]
)
m1 = melb_df['Date'] <= pd.to_datetime('2017-09-01')
m2 = melb_df['Date'] >= pd.to_datetime('2017-05-01')
m = m1 & m2
melb_df[m].groupby('SellerG')['Price'].sum().sort_values(ascending=True)
melb_df.groupby('Rooms')[['Price']].agg('describe')
melb_df.groupby('Rooms')[['Price', 'BuildingArea']].median() # зависимость медианной цены и площади здания от числа комнат
melb_df.groupby(['Rooms', 'Type'])['Price'].mean() #средняя цена в зависимости от числа комнат в разных типах объектов недвижимости

melb_df.groupby(['Rooms', 'Type'])['Price'].mean().unstack() #метод unstack()
melb_df.pivot_table(
    values='Price',    #тоже самое, только метод pivot_table, является основным при создании сводных таблиц
    index='Rooms',
    columns='Type',
    fill_value=0
).round()
melb_df.pivot_table(     #продажи в каждом из регионов в зависимости от того, будний был день или выходной
    values='Price',
    index='Regionname',
    columns='Weekend',
    aggfunc='count'
)

melb_df.pivot_table(  #как зависит средняя и медианная площадь участка (Landsize) от типа объекта (Type) и его региона (Regionname)
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)

melb_df.pivot_table( #
    values='Price',
    index=['Method','Type'],
    columns='Regionname',
    aggfunc='median',
    fill_value=0
)

melb_df.pivot_table( #
    values='Price',
    index='Regionname',
    columns= ['Method', 'Type'],
    aggfunc='median',
    fill_value=0
)

pivot = melb_df.pivot_table(  #результат выполнения метода pivot_table
    values='Landsize',
    index='Regionname',
    columns='Type',
    aggfunc=['median', 'mean'],
    fill_value=0
)
pivot.columns 
#display(pivot['mean']['house'], pivot['mean']['unit'])

#регионы, в которых средняя площадь здания для домов типа house меньше их медианной площади
mask = pivot['mean']['house'] < pivot['median']['house']
filtered_pivot = pivot[mask]
#display(filtered_pivot)
#print(list(filtered_pivot.index))

melb_df.pivot_table(  # зависимость медианной площади (BuildingArea) здания от типа объекта недвижимости (Type) 
    values='BuildingArea',  #и количества жилых комнат в доме (Rooms)
    index='Rooms',
    columns='Type',
    aggfunc='median',
    fill_value=0
)

p = melb_df.pivot_table(  # зависимость медианной цены объекта недвижимости (Price) 
    values='Price',  # от риелторского агентства (SellerG) и типа здания (Type)
    index='SellerG',
    columns='Type',
    aggfunc='median',
    fill_value=0
)
p['unit'].sort_values(ascending=False)
```
